```python
# 2.1
import numpy as np
import random
from math import *
```

```python
quantity = random.randint(7, 11)
list_num1 = [i for i in range(quantity)]
list_num2 = [round(pow(i, e), 2) for i in range(quantity)]
```

```python
# 2.2
a_nums = np.array([list_num1, list_num2])
a_zeros = np.zeros((6, 9))
a_ones = np.zeros((2, 2, 8))
a_one_matrix = np.eye(8)
a_step_matrix = np.arange(10, 1, -1)
```

```python
# 2.5
a1 = np.array(list_num1)
a2 = np.arange(quantity, 0, -1)
a12 = np.array([a1, a2])


square_difference = a1**2 - 2*a1*a2 + a2**2   #квадрат разности
powers = np.where(a1 % 2 == 0, a1 ** e, a1)   # где чётные → i^e, иначе оставляем i
trans_a12 = a12.T #транспонирование
multi_matrix = a1@a2.T #произведение матриц

```

```python
print(a1, a2, a12, square_difference, powers.round(2), trans_a12, multi_matrix, sep="\n")
```

```python
print("\nСтатистики по powers:")
print("среднее =", powers.mean())
print("медиана =", np.median(powers))
print("стд. отклонение =", powers.std().round(3))
print("сумма =", powers.sum())
print("мин / макс =", powers.min(), "/", powers.max())
```
